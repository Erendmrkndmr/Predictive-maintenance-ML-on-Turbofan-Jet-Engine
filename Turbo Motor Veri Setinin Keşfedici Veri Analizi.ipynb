# Korelasyon Isı Tablosu (Korelasyon Analizi)

```python
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize = (58,58))
sns.set(font_scale = 4, font = "Times New Roman")

#Isı tablosunu çizdirme
g = sns.heatmap(df[top_corr_features].corr(),cmap = "RdYlGn", linewidths = 0.1, annot = True, annot_kws = {"size":35})
g.set_xticklabels(g.get_xmajorticklabels(), fontsize = 35)
g.set_yticklabels(g.get_xmajorticklabels(), fontsize = 35)
```

# 0.5'den küçük olan korelasyon değerlerini RUL değerleri ile birlikte tablodan çıkartıyoruz. (Drop)

```python
delete_columns = []

for col in df.columns:
    corr = df[col].corr(df['RUL'])
    if abs(corr) > 0.5:
        pass
    else:
        delete_columns.append(col)
delete_columns
```

```python
new_df = df.drop(columns = delete_columns)
new_df
```

```python
corrmat = new_df.corr()
top_corr_features = corrmat.index
plt.figure(figsize = (58,58))
sns.set(font_scale = 4, font = 'Times New Roman')

# Isı Tablosunu çıkartma
g = sns.heatmap(new_df[top_corr_features].corr(), cmap = "RdYlGn", linewidths = 0.1, annot = True, annot_kws = {"size":35})
g.set_xticklabels(g.get_xmajorticklabels(), fontsize = 35)
g.set_yticklabels(g.get_xmajorticklabels(), fontsize = 35)
```

# Özellik önemi için ekstra ağaç regresyonlar

```python
X = new_df.iloc[:, 0:13]
y = new_df.iloc[:, 13]

from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
print(model.feature_importances_)
```

```python
feat_importances = pd.Series(model.feature_importances_, index = X.columns)
feat_importances.nlargest(13).plot(kind = 'barh', fontsize = 15)
```

```python
X = new_df.iloc[:, 0:-1]
y = new_df.iloc[:, 13]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle = False)

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

from sklearn import metrics
print('RMSE Test Değeri:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

y_pred_train = regressor.predict(X_train)
print('RMSE Eğitim "Trainlenen" Değeri:', np.sqrt(metrics.mean_squared_error(y_train,
                                                                            y_pred_train)))
```

# İleri Özellik Seçimi (Forward Feature Selection)

```python
feat_importances.keys()
```

```python
feat_imp_df = new_df[['Cycle', 'SensorMeasure2', 'SensorMeasure3', 'SensorMeasure4',
       'SensorMeasure7', 'SensorMeasure8', 'SensorMeasure11',
       'SensorMeasure12', 'SensorMeasure13', 'SensorMeasure15',
       'SensorMeasure17', 'SensorMeasure20', 'SensorMeasure21']]
```

```python
feat_imp_df
```

```python
number_of_features = []
test_rmse = []
train_rmse = []

for i in range(1,13):
    number_of_features.append(i)
    
    X = feat_imp_df.iloc[:, 0:i]
    y = new_df.iloc[:, 13]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle = False)
    
    regressor = RandomForestRegressor()
    regressor.fit(X_train, y_train)
    print('Toplam Kullanılan Özellik:', i)
    
    y_pred = regressor.predict(X_test)
    test_rmse.append(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print('Test veri setinde RMSE Değeri:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    
    y_pred_train = regressor.predict(X_train)
    train_rmse.append(np.sqrt(metrics.mean_squared_error(y_train, y_pred_train)))
    print('Eğitim "Training" veri setinde RMSE Değeri:', np.sqrt(metrics.mean_squared_error(y_train, y_pred_train)))
    
    print('\n')
```

```python
plt.figure(figsize = (10,5))
plt.plot(number_of_features, test_rmse)
plt.plot(number_of_features, train_rmse)
plt.xlabel("Özellik Sayısı", fontsize = 18)
plt.ylabel("Karekök Ortalama Hatası", fontsize = 18)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(['Test_rmse', 'Train_rmse'], loc = 'upper right', prop={"size":18})
plt.show()
```
