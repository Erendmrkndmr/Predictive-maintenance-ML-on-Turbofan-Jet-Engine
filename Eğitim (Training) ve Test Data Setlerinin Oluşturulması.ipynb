RUL = EOL - Cycle

* RUL = Kalan Faydalı Ömür (Remaining Useful Life)
* EOL = Hayat Döngüsünün Bitişi (End Of Life Cycles)

```python
EOL = []
for sl_no in ham_data['ID']:
    EOL.append(((ham_data[ham_data['ID'] == sl_no]["Cycle"]).values)[-1])
```

# Test Datasını (Verisini) Hazırlama

```python
delete_columns = ['OpSet1', 'OpSet2',
                  'OpSet3','SensorMeasure1',
                  'SensorMeasure5', 'SensorMeasure6',
                  'SensorMeasure9', 'SensorMeasure10',
                  'SensorMeasure14','SensorMeasure16', 
                  'SensorMeasure18', 'SensorMeasure19']
```

```python
file_no = 4 #Test filelara (1,2,3,4) sl.nosu verdim
###############################
ham_data = pd.read_csv('test_FD00{}.txt'.format(file_no), sep = ' ', header = None)
ham_data = ham_data.drop([26,27], axis = 'columns')
ham_data.columns = ['ID', 'Cycle', 'OpSet1', 'OpSet2', 'OpSet3', 'SensorMeasure1', 'SensorMeasure2', 'SensorMeasure3', 'SensorMeasure4', 'SensorMeasure5', 'SensorMeasure6', 'SensorMeasure7', 'SensorMeasure8', 'SensorMeasure9', 'SensorMeasure10', 'SensorMeasure11', 'SensorMeasure12', 'SensorMeasure13',
                    'SensorMeasure14', 'SensorMeasure15', 'SensorMeasure16', 'SensorMeasure17', 'SensorMeasure18', 'SensorMeasure19', 'SensorMeasure20','SensorMeasure21']
#Burada yine aslında yukarıdaki işleme benzer işlemleri yaptıkm
#Ama bu sefer daha dinamik bir şekilde hazırlamış oldum.
#Örneğin 2,3 veriyide hazırlamak istiyorsak sadece file no'yu değiştirerek kolayca kaydedebiliyoruz.
###############################
given_no_of_cycles = []
for i in range (1, len(ham_data['ID'].unique())+1):
    given_no_of_cycles.append(len(ham_data[ham_data['ID']==i]))
given_no_of_cycles = np.array(given_no_of_cycles).reshape([len(given_no_of_cycles),1])
###############################
cycle_ran_after = pd.read_csv('RUL_FD00{}.txt'.format(file_no), sep = ' ', header = None)
cycle_ran_after = cycle_ran_after.drop([1], axis = 'columns')
cycle_ran_after = np.array(cycle_ran_after)
###############################
EOL = []
for i in ham_data['ID']:
    EOL.append(given_no_of_cycles[i-1]+cycle_ran_after[i-1])
EOL = np.array(EOL).reshape([len(EOL),1])
###############################
ham_data['EOL'] = EOL
ham_data['RUL'] = ham_data['EOL'] - ham_data['Cycle']
ham_data = ham_data.drop(columns = ['EOL'])
ham_data = ham_data.drop(columns = delete_columns)
###############################
```
